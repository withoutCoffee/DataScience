```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/Superstore.csv', encoding='ISO-8859-1')
```

```python
df.columns
```

```python
df.info()
```

Verificando a presença de dados nulos e duplicados no dataset

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

Podemos perceber que não há dados nulos no dataset, portanto não é necessário o uso da função **df.dropna()** e **df.drop_duplicates()** do pandas.

**Alguns insights**

```python
df['Ship Mode'].value_counts().plot(kind='bar',label='Tipos de envio',cmap='bwr')
plt.title('Tipos de envios escolhidos pelos clientes')
plt.show()
```

```python
df['Segment'].value_counts().plot(kind='bar',label='Segmento do cliente',cmap = 'bwr')
plt.title('Segmentos dos clientes')
plt.show()
```

Agora vamos ver que tipo de cliente faz a loja ter mais lucros

```python
df[['Segment','Profit']].groupby('Segment').sum().plot(kind='bar',cmap='bwr')
```

Tipo de produto que mais obtém-se lucro

```python
df[['Category','Profit']].groupby('Category').sum().plot(kind='bar',cmap='bwr')
plt.title('Categorias de produto que mais dão lucro')
plt.show()
```

```python
df[['Sub-Category','Profit']].groupby('Sub-Category').sum().plot(kind='bar',cmap='bwr')
plt.title('Sub-categorias de produtos que mais dão lucro')
plt.show()
```
