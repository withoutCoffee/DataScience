## Análise e Regressão no dataset SuperStore

### 1 - Vasculhando os dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

As libs acima são importantes para analisar, limpar e transformar os dados para que possam ser usados como dados de treinamento para algum algoritmo de machine learning.
- Pandas: Usava para ler e fazer alumas análises sobre os dados, como estatísticas descritivas e aplicação de filtros.
- numpy: Lib de computação numérica, usada para transformação e resolução de problemas relacionados a algebra linear por exemplo.
- Matplotlib: Usava para visualização de dados, com ela podemos extrair algums insights.

```python
df = pd.read_csv('datasets/Superstore.csv', encoding='ISO-8859-1')
```

```python
df.columns
```

```python
print(df.shape)
```

```python
df.info()
```

Verificando a presença de dados nulos e duplicados no dataset

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

Podemos perceber que não há dados nulos no dataset, portanto não é necessário o uso da função **df.dropna()** e **df.drop_duplicates()** do pandas.

**Tratando as datas**

```python
od = df["Order Date"].apply(lambda x: x.replace('/',''))
sd = df["Ship Date"].apply(lambda x: x.replace('/',''))
```

```python
df['Order Date'] = od.apply(lambda x:pd.to_datetime(x,format='%m%d%Y'))
df['Ship Date'] = sd.apply(lambda x:pd.to_datetime(x,format='%m%d%Y'))
```

**Vasculhando os dados**

```python
corr = df.corr()
```

```python
corr.style.background_gradient(cmap='coolwarm')
```

Como a maior parte dos dados são categóricos apenas para algumas colunas é possível aplicar a Correlação de Pearson. A variável com maior correlação com os lucros é Sales, com 0.47, assim tendo um correlação positiva.

**Algumas estatísticas descritivas gerais**

```python
df.describe().style.background_gradient(cmap='coolwarm')
```

Pela visualização do box plot podemos ter uma visão geral da dispersão e assimentria da distribuição dos e dados.

```python
plt.figure(figsize=(20,6))

plt.subplot(1,4,1)
plt.boxplot(df['Sales'])
plt.title('Vendas')

plt.subplot(1,4,2)
plt.boxplot(df['Quantity'])
plt.title('Quantidade do produto')

plt.subplot(1,4,3)
plt.boxplot(df['Discount'])
plt.title('Disconto')

plt.subplot(1,4,4)
plt.boxplot(df['Profit'])
plt.title('Lucro')
```

**Alguns insights**

Gáfico de dispersão, vendas/lucos.

```python
fig, ax = plt.subplots(1,1,figsize=(18,8))
scatter = ax.scatter(df['Sales'], df['Profit'], c=df['Discount'], s=2*df['Quantity']**2, cmap='viridis',alpha=0.65)

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper left", title="Disconto")
ax.add_artist(legend1)

# produce a legend with a cross section of sizes from the scatter
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.5)
legend2 = ax.legend(handles, labels, loc="upper right", title="Quantidade")

plt.xlabel('Vendas')
plt.ylabel('Lucro')
plt.title('Dispersão das vendas pelos lucros')

plt.show()
```

Tipos de envio mais escolhidos pelos clientes

```python
df['Ship Mode'].value_counts().plot(kind='bar',label='Tipos de envio')
plt.title('Tipos de envios escolhidos pelos clientes')
plt.show()
```

**Segmentos dos clientes**

```python
df['Segment'].value_counts().plot(kind='bar',label='Segmento do cliente')
plt.title('Segmentos dos clientes')
plt.show()
```

Agora vamos ver que tipo de cliente faz a loja ter mais lucros

```python
df[['Segment','Profit']].groupby('Segment').sum().plot(kind='bar')
```

**Tipo de produto que mais obtém-se lucro**

```python
df[['Category','Profit']].groupby('Category').sum().plot(kind='bar')
plt.title('Categorias de produto que mais dão lucro')
plt.show()
```

```python
df[['Sub-Category','Profit']].groupby('Sub-Category').sum().sort_values('Profit',ascending=False).plot(kind='bar')
plt.title('Sub-categorias de produtos que mais dão lucro')
plt.show()
```

**Tipos de produtos mais cosumidos a partir do segmento do cliente**

```python
#Consumer
df[df['Segment']=='Consumer']['Sub-Category'].value_counts().plot(kind='bar'),
plt.title('Categorias de produtos mais cosumidos pelos clientes "Consumer"')
plt.show()
```

A partir do gráfico acima podemos perceber que o produto mais consumo é Binders, porém o produto que mais gera lucro é Copiers, assim mesmo que o dataset não tenha o preço de cada produto podemos ter a noção que Copiers são bem mais caras pode conta de seu lucro a partir da venda de menos unidades comparado a outros produtos.

```python
fig, ax = plt.subplots(2,2,figsize=(9,6))
ax[0,0].hist(df['Sales'])
ax[0,0].set_title('Histograma das Vendas')

ax[0,1].hist(df['Quantity'])
ax[0,1].set_title('Histograma das Quantidade de produtos')

ax[1,0].hist(df['Discount'])
ax[1,0].set_title('Histograma do Desconto')

ax[1,1].hist(df['Profit'])
ax[1,1].set_title('Histograma dos Lucros')

plt.show()
```
