**Teste de Heatmap para entender melhor como são as compras ao longo da semana/mês**

```python
ht = df[['Order Date','Sales']].groupby('Order Date').sum()
```

```python
# Criando matrix para heat map mensal

def mheatmap(df,s,e):
    month = np.zeros((s,e))
    for i in range(s*e):
        try:
            #print(df.index[i].week-1)
            #print(month)
            month[df.index[i].week-1,df.index[i].weekday()] = df['Sales'].iloc[i]
            m = i
        except Exception as e:
            break
    return month, m
```

```python
fig, ax = plt.subplots(figsize=(14,8))

matrix,m = mheatmap(ht,4,7)
ax.imshow(matrix)

ax.set_xticks(np.arange(7),labels=np.arange(1,8,1))
ax.set_yticks(np.arange(4),labels=np.arange(1,5,1))

# Loop over data dimensions and create text annotations.
for i in range(4):
    for j in range(7):
        text = ax.text(j, i, matrix[i, j],
                       ha="center", va="center", color="w")

plt.ylabel("Semanas")
plt.xlabel("Dias da Semanas")
plt.xticks(rotation=-90)
plt.title(f"Vedas nos dias da semana do mês {ht.index[m].month} de {ht.index[m].year}")
plt.show()
```
